```python
# titanic data 불러오기
import pandas as pd
base_path = "/content/drive/MyDrive/Colab Notebooks/data/titanic"
train = pd.read_csv(base_path + "/train.csv")
test = pd.read_csv(base_path + "/test.csv")
submission = pd.read_csv(base_path + "/gender_submission.csv")

```

```python
train.info() #age,cabin embarked 에 대해 결측치 처리
```

```python
train[train.isnull().any(axis =1)]
```

### Data Preprocessing

1. 결측치 처리


2. feature selection (분석에 사용하지 않을 column 제거)

```python
# titanic data에서 missing value를 찾아봅니다.
train.dropna()
```

```python
# Embarked column이 NaN인 row를 찾습니다.
train[train.Embarked.isnull()]

#train.Embarked.value_counts() --> "S"로 채우자!

#train[(train.Pclass == 1)&(train.Sex == "female")].Embarked.value_counts() --> "S"로 결측치를 채운다라는 결론

train.loc[train.Embarked.isnull(),"Embarked"] = "S"

train[(train.Pclass == 1)&(train.Sex == "female")].Embarked.value_counts()

train.info()

```

```python
# missing value를 handling 합니다.
# column을 지울까요 / 채울까요?
train.Cabin.value_counts()
#drop 할 컬럼들 골라보기 --> passenger id, name , ticket, cabin
train = train.drop(columns = ["PassengerId","Name","Ticket","Cabin"])
```

```python
train.info()
```

```python
train = train.fillna(train.Age.mean())
train.info()
```

### Feature Engineering

1. Categorical feature encoding

2. Normalization

```python
# categorical feature --> One-hot Encoding, Ordinal Encoding (object 형이 카테고리임. 이걸 변환해주는 작업)

#1. ordinal feature =  ordinary feature 학력,선호도, (어느정도의 순서가 존재한다고 생각하는 것.)
#2. one - hot feature = nominal feature 성별, 부서, 출신학교

train_OHE = pd.get_dummies(train, columns = ["Sex",'Embarked'])
train_OHE
```

```python
# Normalization --> Min-Max scaling

x = train_OHE.drop(columns = "Survived") #input matrix
y = train_OHE.Survived #target vector


from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

### Training

```python
# sklearn에서 배웠던 분류 모델들을 불러와봅니다.
from sklearn.linear_model import SGDClassifier #1. linear classifier
from sklearn.linear_model import LogisticRegression #2. Logistic Regression
from sklearn.tree import DecisionTreeClassifier #3. Decison Tree
from sklearn.ensemble import RandomForestClassifier #4. Random Forest

#평가지표(accuracy 사용하라고 캐글에 나와있어서)
from sklearn.metrics import accuracy_score
```

### Test (Predict)

```python
clf = SGDClassifier()
clf2 = LogisticRegression()
clf3 = DecisionTreeClassifier()
clf4 = RandomForestClassifier()

clf.fit(x,y)
clf2.fit(x,y)
clf3.fit(x,y)
clf4.fit(x,y)

pred = clf.predict(x)
pred2 = clf2.predict(x)
pred3 = clf3.predict(x)
pred4 = clf4.predict(x)
```

```python
print("Linear Classifier ,Accuracy for training : %.4f" % accuracy_score(y,pred))
print("Logistic Regression ,Accuracy for training : %.4f" % accuracy_score(y,pred2))
print("Decision Tree ,Accuracy for training : %.4f" % accuracy_score(y,pred3))
print("Random Forest ,Accuracy for training : %.4f" % accuracy_score(y,pred4))
```
